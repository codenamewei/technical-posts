# Add Videos to Markdown Cells

## Method 1: Display the Videos with HTML Syntax

```python
**Video URL** (Change this to markdown cell to run)

<video src="https://interactive-examples.mdn.mozilla.net/media/cc0-videos/flower.mp4" autoPlay width="500" controls loop>
```

### Parameters to put alongside `<video>` tag
- `width`
    - Example: `width="200"`
- `height`
    - Example: `height="200"`
- `autoPlay`
    - to play automatically: `autoplay="true"`
    - to not play automatically: `autoplay="false"`
- `controls`
    - to include controls: `controls`
    - to not include controls: exclude the controls parameter
- `muted`
    - muted since the beginning, but can toggle by user
    - to mute sounds: `muted`
    - to not mute sounds: exclude the parameter `muted`
- `loop`
    - for the video to restart
- Check out more attributes from [here](https://developer.mozilla.org/en-US/docs/Web/HTML/Element/video#attributes)

# Add Videos to Python Cells

## Method 2: Display the Videos from URL

```python
import mediapy as media
```

```python
video = media.read_video('https://github.com/hhoppe/data/raw/main/video.mp4')
media.show_video(video)
```

## Method 3: Display the Videos from The Local Path

```python
video = media.read_video('metadata/video.mp4')
media.show_video(video, title = "Hello World", fps=5, width=500)
```

## Display multiple videos

```python
videos = [media.read_video('metadata/video.mp4'), media.read_video('metadata/video.mp4')]
media.show_videos(videos)
```

### Optional parameters to put in `show_video` function

- `width`
    - Example: `width="200"`
- `height`
    - Example: `height="200"`
- `fps`
    - Example: `fps=15`
- `title`
    - Example: `title="Hello World"`

- Read more about `show_video` from [here](https://google.github.io/mediapy/mediapy.html#show_video)
- Read more about `show_videos` from [here](https://google.github.io/mediapy/mediapy.html#show_videos)
